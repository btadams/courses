##**Python - Conditional Statement**
**Conditional statements** - in Python are used to make decisions in your code based on certain conditions. They allow you to execute different blocks of code depending on whether a condition is True or False.  The result is a Boolean statement.

Conditional statements are fundamental for controlling the flow of your Python programs. They allow you to write code that responds dynamically to different situations and conditions, making your programs more versatile and useful.
Python supports several types of conditional statements:
<br><br>
The **"if"** statement is the most basic conditional statement. It allows you to execute a block of code only if a specified condition is True.<br><br>
**if condition:**<br>
    \#Code to execut if the condition is true

```python
x = 3
if x < 5: # This is the condition and the start of the 'if' conditional block
    print("x is greater than 5") #This is what happens if the condition is true
```

**if-else Statement:**

The **if-else** statement extends the if statement by allowing you to execute one block of code if the condition is True and another block of code if the condition is **False**.<br><br>
**Note:** There are two blocks of code.  The line starting the block ends with a colon ":"<br><br>
**if condition:**<br>
    \# Code to execute if the condition is True<br>
**else:**<br>
    \# Code to execute if the condition is False

```python
x = 3
if x > 5:
    print("x is greater than 5")
else:

    print("x is not greater than 5")
print('This represents the line of code following the "if" condition')
```

**if-elif-else Statement (Multiple Conditions):**

The **if-elif-else** statement allows you to test multiple conditions in sequence. It will execute the first block of code whose condition is True, or the last else block if none of the earlier conditions are True.<br><br>
**if condition1:**<br>
    \# Code to execute if condition1 is True<br>
**elif condition2:**<br>
    \# Code to execute if condition2 is True<br>
**else:**<br>
    \# Code to execute if neither condition1 nor condition2 is True


```python
x = 10
if x > 15:
    print("x is greater than 15")
elif x > 5:
    print("x is greater than 5 but not greater than 15")
else:
    print("x is not greater than 5")
```

**Nested if Statements:**

You can nest if statements inside other if, elif, or else blocks to create more complex decision-making logic.

```python
x = 4
if x > 5:
    print("x is greater than 5")
    if x > 15:  # This is the True part of the partent if statment
        print("x is also greater than 15")  # This is the True part of the Child if statment
    else:
        print("x is not also greater than 15") # This is the False part of the Child if Statement
else:
  print("x is less than 5")
```

##**Python Loops**
Looping in Python, or in programming in general, allows you to perform repetitive tasks and automate processes. Such as:
1. Repeat a Block of Code
2. Iterate Over a Sequence
3. Preform Calculations
4. Read and Process Data
5. Control Flow
6. Handle Collections
7. Implement Algorithms
8. Create Inteactive programs

In Python, there are two main types of loops: **for loops and while loops**. These loops allow you to repeatedly execute a block of code as long as a certain condition is met

**for Loop:**

A **for** loop is used for iterating over a sequence (array) (such as a list, tuple, string, or range) and executing a block of code for each item in the sequence (array).

**for variable in sequence:**<br>
    \# Code to execute for each item in the sequence

```python
fruits = ["apple", "banana", "cherry"]  #This is a list (sequence item/array). It defines our collection
for fruit in fruits: # "fruit" is the temporary variable that will will represent eache piece of fruit in the list
    print("Fun Fruit:",fruit)
```

***while Loop:**

A while loop is used to repeatedly execute a block of code as long as a specified condition is **True**. The loop continues until the condition becomes **False**.

**while condition:**<br>
    \# Code to execute as long as the condition is True

```python
count = 1 # count represents an initial value
while count <= 5: #T his is a boolean expression that is tested before the block of code executes
    print("The count is:",count)
    count += 1 # This line enumerates the "count" value by 1
```

```python
# Nested Loops
# You can nest one type of loop inside another to create more complex looping patterns.
# Example of a nested for loop:
for i in range(3): #Range is a sequence function that allows you to define a range (sequence) of numbers (Start - Stop Before - Step)
    for j in range(3):
        print(f"({i}, {j})")
```

```python
# Example of a nested while loop:
i = 1
while i <= 3:
    j = 1
    while j <= 3:
        print(f"({i}, {j})")
        j += 1
    i += 1
```

**Loop Control Statements:**

Python provides loop control statements like **break**, **continue**, and **else** that allow you to customize the behavior of loops.

**break:** Terminates the loop prematurely.<br>
**continue:** Skips the current iteration and moves to the next one.<br>
**else with loops:** You can use else with a loop to specify code that should always run after the loop has finished executing, but only if the loop completed without encountering a break.**else** works with the "for" and "while" loops

```python
# Break
for i in range(1, 6):
    if i == 3:
        break
    print(i)
```

```python
# Continue
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

```python
# else
for i in range(1, 6):
    if i == 7:
        break
    print(i)
else:
    print("Loop completed without encountering a break.")
```

#**Python - Basic Data Structures**
Python Data Structures are represented by sequence (array) objects.

These data structures have a level of Mutability, meaning some are mutable and some are immutable.

Mutability refers to whether an object's state (the data it holds) can be modified after the object is created.

**Mutable** - objects are those whose state can be changed after they are created. You can add, remove, or modify elements or properties of mutable objects.

**Immutable** - objects are those whose state cannot be changed after they are created. Once an immutable object is created, it cannot be modified. If you want to change an immutable object's value, you must create a new object.

***Implications of Mutability:***

*Mutability has implications for code behavior and safety. When you modify a mutable object, it may affect other parts of your code that reference the same object. This can lead to unexpected behavior.*

Immutable objects are safer when dealing with shared data because they cannot be modified. Any operation that seems to modify an immutable object actually creates a new object.

##**Range - range([start],[stop before],[step])**
1. **start (optional):** The starting value of the range (inclusive). If omitted, the range starts at 0.
2. **stop before:** The stopping value of the range (exclusive). The range will generate values up to, but not including, this value.
3. **step (optional):** The step size between values in the range. If omitted, the default step size is 1.


```python
# Range
my_range = range(5)
print("my_range:",list(my_range))  # Output: [0, 1, 2, 3, 4]

# Using the Start - Stop before and - Step
my_range = range(1, 10, 2)
print("my_range with the [start at 1] [stop before 10] [step by2]:",list(my_range))  # Output: [1, 3, 5, 7, 9]
```

##**Python - List Comprehensions**
**List comprehensions** in Python provide a concise and readable way to create lists. They allow you to generate a new list by applying an expression to each item in an iterable (e.g., a list, tuple, or range) and optionally filtering items based on a condition.<br>The compreshension combines the 'for' statement with the automatic creation of a list and the possibility of an 'if' statment.<br> List comprehensions are a powerful tool for simplifying code that involves iterating over data and transforming it into a new list.

**new_list = [expression for item in iterable if condition]**

```python
# Creating a list of squares
squares = [x**2 for x in range(1, 6)]
print("The list of squares:",squares)  # Output: [1, 4, 9, 16, 25]
```

```python
# A list comprehension including an 'if' statement
evens = [x for x in range(1, 11) if x % 2 == 0]  #We used '%' (Mod) to test if the remainder is equal (==) to 0
print("The even number are:",evens)  # Output: [2, 4, 6, 8, 10]
```

```python
# Transforming and Filtering a List
words = ["apple", "banana", "cherry", "date", "fig"]
print("The 'words' list:",words)
word_lengths = [f'length of {word} is {len(word)}' for word in words if len(word) > 3] #This expression uses string-interpolation
print(word_lengths)  # Output: [5, 6, 4]
```

####**Writing to a File**<br>
**Write Mode ('w'):**

When a file is opened in **write mode ('w')***, it is writable.
If the file exists, its previous content will be truncated (deleted), and you can write new data to it.
If the file doesn't exist, a new empty file is created, and you can write data to it.

**Append Mode ('a'):**

When a file is opened in **append mode ('a')**, it is also writable.
If the file exists, new data will be added to the end of the existing content without overwriting it.
If the file doesn't exist, a new empty file is created, and you can append data to it.

**Write Binary Mode ('wb'):**

When a file is opened in write **binary mode ('wb')**, it is writable for binary data.
This mode is typically used for writing binary files like images or non-text files.

The writability of a file also depends on file permissions and the directory where the file is located. If you don't have write permissions for a file or the directory, you won't be able to write to the file even if you open it in a writable mode. Always ensure you have the necessary permissions to write to a file or directory before attempting to write data.

```python
# The Answer to the Mystery
# Riddle: I am a number between 1 and 10. I am even and greater than 5. What am I?

# Your task is to write Python code to find and print the answer to the riddle.
# You should print the number that satisfies the riddle's conditions.

# Hint: Use an if statement to check the conditions and print the answer.

# Write your code below this comment:

# Iterate through numbers from 1 to 10
for number in range(1, 11):
    # Check if the number is even and greater than 5
    if number % 2 == 0 and number > 5:
        # Print the answer
        print("The answer is:", number)
```
